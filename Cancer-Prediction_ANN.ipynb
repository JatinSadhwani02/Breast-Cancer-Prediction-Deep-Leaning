```python
def correlation(data , threshold):
    corr = data.corr()['diagnosis'].sort_values(ascending=False)[1:]
    abs_corr = abs(corr)
    relevant_features = abs_corr[abs_corr<threshold]
    return relevant_features
```

```python
corr_features = correlation(data,0.45)
```

```python
data.drop(columns=['smoothness_worst','symmetry_worst','texture_mean','concave points_se','smoothness_mean','symmetry_mean',
                  'fractal_dimension_worst','compactness_se','concavity_se','fractal_dimension_se','id','symmetry_se','texture_se',
                  'fractal_dimension_mean','smoothness_se'],axis=1,inplace=True)
```

## show shape after delete features

## Divide dataset into dependent and independent features

```python
X = data.drop('diagnosis',axis=1)
y = data['diagnosis']
```

```python
X.shape , y.shape
```

## split data into train and test

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,y,test_size=0.1,random_state=42)
```

```python
X_train.shape , X_test.shape , Y_train.shape , Y_test.shape
```

## Feature Scaling

```python
sc = MinMaxScaler()
X_train_mms = sc.fit_transform(X_train)
X_test_mms = sc.transform(X_test)
```

```python
(X_train_mms.min() , X_train_mms.max())
```

```python
(X_test_mms.min() , X_test_mms.max())
```

## Define model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
model.add(Dense(units=12,activation='relu',input_dim=16,kernel_initializer="he_uniform"))
model.add(Dense(units=12,activation='relu',kernel_initializer="he_uniform"))
model.add(Dense(units=1,activation='sigmoid',kernel_initializer="glorot_uniform"))
```

```python
model.compile(loss="binary_crossentropy",metrics=['accuracy'],optimizer='adam')
```

```python
model.fit(X_train_mms,Y_train , validation_data=(X_test_mms,Y_test),epochs=300,batch_size=16)
```

```python
len(X_test_mms)
```

## Prediction with test dataset

```python
pred = model.predict_classes(X_test_mms)
```

```python
pred
```

```python
result = pred.tolist()
result = ["Malignant" if p[0]==1 else "Benign" for p in result ]
actual = ["Malignant" if p==1 else "Benign" for p in Y_test]
```

## check actual and predicted value

```python
pd.DataFrame(np.c_[actual , result],columns=['Actual','Predicted'])
```

## Check accuracy ,  Confusion matrix and Classification report

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(pred,Y_test)
```

```python
confusion_matrix(pred,Y_test)
```

```python
print(classification_report(pred,Y_test))
```
